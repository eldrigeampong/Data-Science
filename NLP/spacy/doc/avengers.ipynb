# The Avengers

```python
import spacy
import os
from spacy import displacy
```

```python
# get current working directory
cwd = os.getcwd()

filepath = os.path.join(cwd, 'files', 'avengers.txt')

# read avengers text
with open(filepath, 'r') as f:
    text = f.read()
    print(text)
```

```python
print(len(text))
```

```python
for idx, string in enumerate(text):
    print(idx, string)
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
doc = nlp(text)
print(doc)
```

```python
print(len(doc))                                                                     # get the number of tokens in the document
```

```python
assert doc[1].text == 'film'                                                        # get the token at index position 1
assert doc[377].text == '.'                                                         # get the token at index position 377
assert doc[3:11].text == 'development began when Marvel Studios received a loan'    # get tokens between index positions 3 and 11
```

```python
[t.text for t in doc[20:25]]                                                        # iterate over tokens between index positions 20 and 25
```

```python
for idx, token in enumerate(doc):                                                   # return the index of each token in the text
    print(idx, token)
```

```python
from spacy.tokens import Span

doc.set_ents([Span(doc=doc, start=137, end=139, label='GPE')])                 # set the named entity for span between 137 and 139 as Geopolitical entity
ents = list(doc.ents)

print(f'entity name: {ents[0].text}')
print(f'entity label: {ents[0].label_}')
```

```python
# make a semantic similarity estimate
text_sim1 = doc[167:178]
text_sim2 = doc[205:252]
text_sim3 = doc[252:312]
text_sim4 = doc[312:341]

print(text_sim1)
print('\n', text_sim2)
print('\n', text_sim3)
print('\n', text_sim4)
```

```python
textsim1_textsim2 = text_sim1.similarity(text_sim2)                           # similarity between text 1 and 2
print('Simmilarity: {0:.2%}'.format(textsim1_textsim2))
```

```python
textsim3_textsim4 = text_sim3.similarity(text_sim4)                           # similarity between text 3 and 4
print('Simmilarity: {0:.2%}'.format(textsim3_textsim4))
```
