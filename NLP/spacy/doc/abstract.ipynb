Source: https://www.science.org/doi/abs/10.1126/science.aaa8415

```python
import os
import spacy
from spacy import displacy
```

# Read Text Document

```python
# get current working directory
cwd = os.getcwd()

fullfilepath = os.path.join(cwd, 'files', 'abstract.txt')

with open(fullfilepath, 'r') as f:
    text = f.read()
    print(text)
```

# Data Exploration

```python
# load spacy instance
nlp = spacy.load('en_core_web_md')
```

```python
# convert text file to spacy doc object
doc = nlp(text)
```

```python
for token in doc.sents:
    print(token, end='\n')
```

```python
print(f'Text length: {len(doc)}')

for idx, token in enumerate(doc):
    print(idx, token.text, token.pos_, token.dep_)
```

# Document Visualization

```python
displacy.render(doc, style='dep')
```
