# Grouping and Aggregation

Filter By Region

```python
regions = ['Africa', 'Northern Africa', 'Sub-Saharan Africa', 'Eastern Africa', 'Middle Africa', 'Southern Africa', 'Western Africa', 
           'Americas', 'Northern America', 'Latin America & the Caribbean', 'Caribbean', 'Central America', 'South America', 'Asia', 
           'Central Asia', 'Eastern Asia', 'South-central Asia', 'South-eastern Asia', 'Southern Asia', 'Western Asia', 'Europe', 
           'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Western Europe', 'Oceania', 'Australia and New Zealand', 'Melanesia']

df_reg_filt = df1[df1['Region/Country/Area'].isin(regions)]
df_reg_filt
```

```python
df_reg_filt.rename(columns={
    'Region/Country/Area': 'Region'
}, inplace=True)

df_reg_filt
```

Filter GDP in Millions

```python
millions = ['GDP in current prices (millions of US dollars)', 
            'GDP in constant 2010 prices (millions of US dollars)']


df_reg_filt_mill = df_reg_filt[df_reg_filt['Series'].isin(millions)]
df_reg_filt_mill
```

Aggregated GDP Grouped By Region in Millions of US Dollars

```python
df_reg_gpd_agg = df_reg_filt_mill.groupby('Region').agg(
    min_GDP = pd.NamedAgg('Value', np.min),
    mean_GDP = pd.NamedAgg('Value', np.mean),
    max_GDP = pd.NamedAgg('Value', np.max),
    total_GDP = pd.NamedAgg('Value', np.sum)
)

df_reg_gpd_agg
```

Filter Top and Bottom 10 Regions GDP in Millions of Dollars

```python
top_10_gdp_reg = df_reg_gpd_agg['total_GDP'].sort_values(ascending=False).head(10)
bottom_10_gdp_reg = df_reg_gpd_agg['total_GDP'].sort_values(ascending=True).head(10)

print(top_10_gdp_reg)
print('\n')
print(bottom_10_gdp_reg)
```

# Data Visualization By Region

```python
# helper function for formatting plot y-axis
from matplotlib.ticker import FuncFormatter

def million(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

formatter = FuncFormatter(million)


# helper function to plot horizontal bar chart 
def plot_hbar(y, width, align='center', xlabel=None, ylabel=None, title=None, color=None):

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()

    # plot bar chart
    ax.barh(
        y=y, width=width,
        align=align
    )

    # Add informative axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    # labels read top-to-bottom
    ax.invert_yaxis()  

    # set axis formatter
    ax.xaxis.set_major_formatter(formatter)

    # despine x-axis and y-axis
    sns.despine(left=True, bottom=True)  
```

Top 10 Regions

```python
plot_hbar(y=top_10_gdp_reg.keys(), width=top_10_gdp_reg.values, 
          xlabel='GDP in current prices ($)', ylabel='Region', title='GDP by Region')
```

Bottom 10 Regions

```python
plot_hbar(y=bottom_10_gdp_reg.keys(), width=sorted(bottom_10_gdp_reg.values), 
          xlabel='GDP in current prices ($)', ylabel='Region', title='GDP by Region')
```

```python
# helper function for formatting plot y-axis
from matplotlib.ticker import FuncFormatter

def thousand(x, pos):
    'The two args are the value and tick position'
    return '%1.1fK' % (x * 1e-3)

formatter1 = FuncFormatter(thousand)


# helper function to plot horizontal bar chart 
def plot_hbar1(y, width, align='center', xlabel=None, ylabel=None, title=None, color=None):

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()

    # plot bar chart
    ax.barh(
        y=y, width=width,
        align=align
    )

    # Add informative axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    # labels read top-to-bottom
    ax.invert_yaxis()  

    # set axis formatter
    ax.xaxis.set_major_formatter(formatter1)

    # despine x-axis and y-axis
    sns.despine(left=True, bottom=True)  
```

# GDP real rates of growth in Percent by Region Exploration

Get Continents Dataframe

```python
africa_df = df_reg_gdp_rates.get_group('Africa')
americas_df = df_reg_gdp_rates.get_group('Americas')
asia_df = df_reg_gdp_rates.get_group('Asia')
anz_df = df_reg_gdp_rates.get_group('Australia and New Zealand')
oceania_df = df_reg_gdp_rates.get_group('Oceania')
europe_df = df_reg_gdp_rates.get_group('Europe')
```

```python
# bar plot helper function
def plot_bar(df, x, y, xlabel=None, ylabel=None, title=None):

    # set plot theme
    sns.set_theme(style='white', context='talk')

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar chart
    sns.barplot(
        x=x, y=y, 
        data=df, ci=None,
        ax=ax
    )

    sns.despine(left=True, bottom=True)

    # matplotlib axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    return plot_bar
```

```python
plot_bar(df=africa_df, 
         x='Year', y='Value', 
         xlabel='Year', ylabel='GDP real rates of growth (percent)', title='GDP Real Rates of Growth (Africa)')
```
