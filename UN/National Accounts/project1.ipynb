# Data Visualization

```python
from matplotlib.ticker import FuncFormatter

# y-axis formatter helper functions
def million(x, pos):
    return '%1.1fM' % (x * 1e-6)

format_mill = FuncFormatter(million)


def thousand(x, pos):
    return '%1.1fK' % (x * 1e-3)

format_tsd = FuncFormatter(thousand)


def percentage(x, pos):
    return '%.1f%%' % (x * 1)

format_pct = FuncFormatter(percentage)


# line plot helper function
def plot_line(data, x, y, hue=None, size=None, style=None, title=None, xlabel=None, ylabel=None, formatter=None):
    # create figure and axis
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot line chart
    sns.lineplot(
        data=data, 
        x=x, y=y,
        hue=hue, size=size, style=style,
        ax=ax)

    # label plot axes
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)

    # set y-axis formatter
    ax.yaxis.set_major_formatter(formatter=formatter)

    # despine plot axes
    sns.despine(bottom=True, left=True)

    return plot_line
```

# GDP in current prices exploration

```python
plot_line(
    data=df1_grpd.get_group(('Germany', 'GDP in current prices (millions of US dollars)')),
    x='Year', y='Value',
    hue='Series',
    title='Germany', xlabel='Year', ylabel='Value',
    formatter=format_mill
)
```

```python
plot_line(
    data=df1_grpd.get_group(('China', 'GDP in current prices (millions of US dollars)')),
    x='Year', y='Value',
    hue='Series',
    title='China', xlabel='Year', ylabel='Value',
    formatter=format_mill
)
```
