```python
df1 = df.dropna()                                                                   # drop missing values
df1
```

# Grouping and Aggregation

```python
country_area_grouped = df1.groupby('Country_or_Area')
country_area_grouped.head()
```

# Aggregation by Trade (USD)

```python
agg_trade_stats = country_area_grouped[['Country_or_Area', 'Year', 'Flow', 'Trade (USD)']].agg(
    min_trade = pd.NamedAgg('Trade (USD)', np.min),
    mean_trade = pd.NamedAgg('Trade (USD)', np.mean),
    max_trade = pd.NamedAgg('Trade (USD)', np.max),
    total_trade = pd.NamedAgg('Trade (USD)', np.sum),
)

agg_trade_stats
```

```python
agg_trade_stats['total_trade'].sort_values(ascending=False).head(10)                 # top ten country or area with the highest trade in dollars
```

```python
top_10_trade = agg_trade_stats['total_trade'].sort_values(ascending=False).head(10)

fig, ax = plt.subplots()

ax.barh(top_10_trade.keys(), top_10_trade.values, align='center')
ax.invert_yaxis()                                                       # labels read top-to-bottom
ax.set(xlabel='Trade ($)', ylabel='Country or Area', title='Trade of Goods (Meat and edible meat offal)')
```

```python
agg_trade_stats['total_trade'].sort_values(ascending=True).head(10)           # top ten country or area with the lowest trade in dollars
```

```python
low_10_trade = agg_trade_stats['total_trade'].sort_values(ascending=True).head(10)

fig, ax = plt.subplots()

hbars = ax.barh(low_10_trade.keys(), low_10_trade.values, align='center')
ax.set(xlabel='Trade ($)', ylabel='Country or Area', title='Trade of Goods (Meat and edible meat offal)')
```

# Aggregation by Weight (kg)

```python
agg_wt_stats = country_area_grouped[['Country_or_Area', 'Year', 'Flow', 'Weight (kg)']].agg(
    min_wt = pd.NamedAgg('Weight (kg)', np.min),
    mean_wt = pd.NamedAgg('Weight (kg)', np.mean),
    max_wt = pd.NamedAgg('Weight (kg)', np.max),
    total_wt = pd.NamedAgg('Weight (kg)', np.sum),
)

agg_wt_stats
```

```python
agg_qty_stats = country_area_grouped[['Country_or_Area', 'Year', 'Flow', 'Quantity']].agg(
    min_qty = pd.NamedAgg('Quantity', np.min),
    mean_qty = pd.NamedAgg('Quantity', np.mean),
    max_qty = pd.NamedAgg('Quantity', np.max),
    total_qty = pd.NamedAgg('Quantity', np.sum),
)

agg_qty_stats
```
