# Data Visualization

```python
from matplotlib.ticker import FuncFormatter

# helper function to round date to the nearest whole number
def x_date_format(x, pos):
    return '%1.0f' % (round(x))

date_format = FuncFormatter(x_date_format)


# line plot helper function
def plot_line(data, x, y, hue=None, ci=None, xlabel=None, ylabel=None, title=None):
    # create figure and axis
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot line chart
    sns.lineplot(
        data=data,
        x=x, y=y, hue=hue, ci=ci,
        ax=ax 
    )

    # plot label
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)

    # x-axis formatter
    ax.xaxis.set_major_formatter(date_format)

    return plot_line
```

```python
plot_line(
    data=df1_grpd.get_group('Germany'),
    x='Year', y='Value', ci=None, 
    xlabel='Year', ylabel='Percentage of Internet Usage', title='Germany'
)
```
