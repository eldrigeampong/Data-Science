# Data Visualization

```python
from matplotlib.ticker import FuncFormatter

# y-axis formatter functions
def million(x, pos):
    return '%1.1fM' % (x * 1e-6)

format_mill = FuncFormatter(million)


def thousand(x, pos):
    return '%1.1fK' % (x * 1e-3)

format_tsd = FuncFormatter(thousand)


# bar plot helper function
def plot_bar(data, x, y, ci=None, hue=None, color=None, title=None, xlabel=None, ylabel=None, formatter=None):
    # create figure and axis
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar charts
    sns.barplot(
        data=data,
        x=x, y=y, 
        ci=ci, hue=hue, color=color,
        ax=ax)

    # label plot axes
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)

    # y-axis formatter
    ax.yaxis.set_major_formatter(formatter=formatter)

    # despine plot axes
    sns.despine(bottom=True, left=True)

    return plot_bar


# pie chart helper function
def plot_pie(data, x, labels, explode: list, startangle: int, shadow: bool, title=None, xlabel=None, ylabel=None):
    # create figure and axis
    fig1, ax1 = plt.subplots()

    # plot pie chart
    ax1.pie(data=data, x=x, explode=explode, labels=labels, autopct='%1.1f%%', shadow=shadow, startangle=startangle)

    # label plot axes
    ax1.set(title=title, xlabel=xlabel, ylabel=ylabel)

    return plot_pie
```

# International migrant stock: Both sexes (number)

```python
plot_bar(
    data=df1_grpd.get_group(('Total, all countries or areas', 'International migrant stock: Both sexes (number)')),
    x='Year', y='Value',
    title='Total, all countries or areas', xlabel='Year', ylabel='International migrant stock: Both sexes (number)',
    formatter=format_mill
)
```

# International migrant stock: Both sexes (% total population)

```python
plot_pie(
    data=df1_grpd.get_group(('Africa', 'International migrant stock: Both sexes (% total population)')),
    x='Value', explode=[0, 0, 0, 0], labels='Year',
    startangle=90, shadow=True,
    title='Africa', xlabel='International migrant stock: Both sexes (% total population)'
)
```
