# Grouping and Aggregation

Filter By Region

```python
regions = ['Africa', 'Northern Africa', 'Sub-Saharan Africa', 'Eastern Africa', 'Middle Africa', 'Southern Africa', 'Western Africa', 
           'Northern America', 'Latin America & the Caribbean', 'Caribbean', 'Latin America', 'Central America', 'South America', 
           'Asia', 'Central Asia', 'Eastern Asia', 'South-central Asia', 'South-eastern Asia', 'Southern Asia', 'Western Asia', 
           'Europe',  'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Western Europe', 'Oceania', 'Australia and New Zealand', 
           'Melanesia']

df_reg_filt = df1[df1['Region/Country/Area'].isin(regions)]
df_reg_filt
```

```python
df_reg_filt.rename(columns={
    'Region/Country/Area': 'Region'
}, inplace=True)

df_reg_filt
```

Filter International migrant stock: Both sexes (number)

```python
number = ['International migrant stock: Both sexes (number)']

df_reg_filt_ims = df_reg_filt[df_reg_filt['Series'].isin(number)]
df_reg_filt_ims
```

Aggregated International migrant stock

```python
df_reg_gpd_agg = df_reg_filt_ims.groupby('Region').agg(
    min_ims = pd.NamedAgg('Value', np.min),
    mean_ims = pd.NamedAgg('Value', np.mean),
    max_ims = pd.NamedAgg('Value', np.max),
    total_ims = pd.NamedAgg('Value', np.sum)
)

df_reg_gpd_agg
```

Filter Top and Bottom 10 Regions International migrant stock

```python
top_10_ims_reg = df_reg_gpd_agg['total_ims'].sort_values(ascending=False).head(10)
bottom_10_ims_reg = df_reg_gpd_agg['total_ims'].sort_values(ascending=True).head(10)

print(top_10_ims_reg)
print('\n')
print(bottom_10_ims_reg)
```

# Data Visualization By Region

```python
# helper function for formatting plot y-axis
from matplotlib.ticker import FuncFormatter

def million(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

formatter = FuncFormatter(million)


# helper function to plot horizontal bar chart 
def plot_hbar(y, width, color=True, align='center', xlabel=None, ylabel=None, title=None):

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()

    # plot bar chart
    ax.barh(
        y=y, width=width,
        color='b',
        align=align
    )

    # Add informative axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    # labels read top-to-bottom
    ax.invert_yaxis()  

    # set axis formatter
    ax.xaxis.set_major_formatter(formatter)

    # despine x-axis and y-axis
    sns.despine(left=True, bottom=True)  
```

Top 10 Regions

```python
plot_hbar(y=top_10_ims_reg.keys(), width=top_10_ims_reg.values, 
          xlabel='International migrant stock', ylabel='Region', title='International migrant stock by Region')
```

Bottom 10 Regions

```python
plot_hbar(y=bottom_10_ims_reg.keys(), width=bottom_10_ims_reg.values, 
          xlabel='International migrant stock', ylabel='Region', title='International migrant stock by Region')
```

Filter Africa, Americas, Asia, Australia and New Zealand, Oceania and Europe Continents

```python
# bar plot helper function
def plot_bar(df, x, y, hue, xlabel=None, ylabel=None, title=None):

    # set plot theme
    sns.set_theme(style='white', context='talk')

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar chart
    sns.barplot(
        x=x, y=y, 
        hue=hue, data=df, ci=None, 
        ax=ax
    )

    # despine axes
    sns.despine(left=True, bottom=True)

    # matplotlib axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    return plot_bar
```

```python
plot_bar(df=continents_df,
         x='Year', y='Value', 
         hue='Region', 
         xlabel='Year', ylabel='International migrant stock')
```

# Total refugees and people in refugee-like situations by Continent Exploration

```python
# helper function for formatting plot y-axis
from matplotlib.ticker import FuncFormatter

def million(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)

formatter1 = FuncFormatter(million)


def plot_bar1(df ,x, y, hue=True, xlabel=None, ylabel=None, title=None):

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # create barplot
    sns.barplot(
    x=x, y=y,
    hue='Region', ci=None,
    data=df, ax=ax)

    # label axis
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    # Add informative axis label
    ax.yaxis.set_major_formatter(formatter1)

    # despine x-axis and y-axis
    sns.despine(left=True, bottom=True)

    # add plot legend
    ax.legend(loc=9)

    return plot_bar1
```

```python
continents = ['Africa', 'Americas', 'Asia', 'Australia and New Zealand', 'Oceania', 'Europe']

continents_df1 = df_reg_filt_totref[df_reg_filt_totref['Region'].isin(continents)]
continents_df1.head()
```

```python
plot_bar1(
    df=continents_df1,
    x='Year', y='Value',
    xlabel='Year', ylabel='Total refugees and people in refugee-like situations',
)
```
