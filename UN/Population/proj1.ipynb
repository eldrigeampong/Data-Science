# Grouping and Aggregation

Filter By Region

```python
regions = ['Africa', 'Northern Africa', 'Sub-Saharan Africa', 'Eastern Africa', 'Middle Africa', 'Southern Africa', 'Western Africa', 
           'Americas', 'Northern America', 'Latin America & the Caribbean', 'Caribbean', 'Central America', 'South America', 'Asia', 
           'Central Asia', 'Eastern Asia', 'South-central Asia', 'South-eastern Asia', 'Southern Asia', 'Western Asia', 'Europe', 
           'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Western Europe', 'Oceania', 'Australia and New Zealand', 'Melanesia']

df_reg_filt = df1[df1['Region/Country/Area'].isin(regions)]
df_reg_filt
```

```python
df_reg_filt.rename(columns={
    'Region/Country/Area': 'Region'
}, inplace=True)

df_reg_filt
```

Filter Population in Millions

```python
millions = ['Population mid-year estimates (millions)',
            'Population mid-year estimates for males (millions)',
            'Population mid-year estimates for females (millions)']

df_reg_filt_mill = df_reg_filt[df_reg_filt['Series'].isin(millions)]
df_reg_filt_mill
```

```python
df_reg_filt_mill['Value'] = df_reg_filt_mill['Value'] * 1e6                     # multiple values series by 10^6
df_reg_filt_mill
```

Aggregated Population mid-year estimates in Millions of US Dollars

```python
df_reg_gpd_agg = df_reg_filt_mill.groupby('Region').agg(
    min_pop = pd.NamedAgg('Value', np.min),
    mean_pop = pd.NamedAgg('Value', np.mean),
    max_pop = pd.NamedAgg('Value', np.max),
    total_pop = pd.NamedAgg('Value', np.sum)
)

df_reg_gpd_agg
```

Filter Top and Bottom 10 Regions Population mid-year estimates in Millions of Dollars

```python
top_10_pop_reg = df_reg_gpd_agg['total_pop'].sort_values(ascending=False).head(10)
bottom_10_pop_reg = df_reg_gpd_agg['total_pop'].sort_values(ascending=True).head(10)

print(top_10_pop_reg)
print('\n')
print(bottom_10_pop_reg)
```

# Data Visualization By Region

```python
# helper function for formatting plot y-axis
from matplotlib.ticker import FuncFormatter

def billion(x, pos):
    'The two args are the value and tick position'
    return '%1.1fBn' % (x * 1e-9)

formatter = FuncFormatter(billion)


# helper function to plot horizontal bar chart 
def plot_hbar(y, width, color=True, align='center', xlabel=None, ylabel=None, title=None):

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()

    # plot bar chart
    ax.barh(
        y=y, width=width,
        color='b',
        align=align
    )

    # Add informative axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    # labels read top-to-bottom
    ax.invert_yaxis()  

    # set axis formatter
    ax.xaxis.set_major_formatter(formatter)

    # despine x-axis and y-axis
    sns.despine(left=True, bottom=True)  
```

Top 10 Regions

```python
plot_hbar(y=top_10_pop_reg.keys(), width=top_10_pop_reg.values, 
          xlabel='Population mid-year estimates', ylabel='Region', title='Population mid-year estimates by Region')
```

Bottom 10 Regions

```python
plot_hbar(y=bottom_10_pop_reg.keys(), width=bottom_10_pop_reg.values, 
          xlabel='Population mid-year estimates', ylabel='Region', title='Population mid-year estimates by Region')
```

Filter Africa, Americas, Asia, Australia and New Zealand, Oceania and Europe Continents

```python
continents = ['Africa', 'Americas', 'Asia', 'Australia and New Zealand', 'Oceania', 'Europe']

continents_df = df_reg_filt_sr[df_reg_filt_sr['Region'].isin(continents)]
continents_df.head()
```

```python
# bar plot helper function
def plot_bar(df, x, y, hue, xlabel=None, ylabel=None, title=None):

    # set plot theme
    sns.set_theme(style='white', context='talk')

    # Initialize the matplotlib figure
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar chart
    sns.barplot(
        x=x, y=y, 
        hue=hue, data=df, ci=None, 
        ax=ax
    )

    # despine axes
    sns.despine(left=True, bottom=True)

    # matplotlib axis label
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)

    return plot_bar
```

```python
plot_bar(df=continents_df,
         x='Year', y='Value', 
         hue='Region', 
         xlabel='Year', ylabel='Sex ratio (males per 100 females)')
```
