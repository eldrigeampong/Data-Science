# Data Visualization

```python
# bar plot helper function
def plot_bar(data, x, y, ci=None, hue=None, color=None, title=None, xlabel=None, ylabel=None):
    # create figure and axis
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar chart
    sns.barplot(
        data=data, 
        x=x, y=y, 
        ci=ci, hue=hue, color=color,
        ax=ax)

    # label plot axes
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)

    # despine axis
    sns.despine(bottom=True, left=True)

    return plot_bar
```

# Major Trading Partner Exploration (Countries)

```python
plot_bar(
    data=df1_gpd.get_group('Ghana'),
    x='Year', y='Value', ci=None,
    hue='MajorTradingPartner_1',
    title='Ghana', xlabel='Year', ylabel='Value (%)'
)
```
