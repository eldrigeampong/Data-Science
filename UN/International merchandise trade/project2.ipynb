Helper function to visualize individual countries or areas

```python
# helper function to plot bar charts
def plot_bar(x, y, hue, df, main_title=None, xlabel=None, ylabel=None, leg_title=None, loc=None, leg_labels=None):
    # initiate matplot figure
    fig, ax = plt.subplots()
    fig.autofmt_xdate()

    # plot bar chart
    sns.barplot(
        x=x, y=y,
        hue=hue, ci=None,
        data=df)

    # label plot axis
    ax.set(title=main_title, xlabel=xlabel, ylabel=ylabel)

    # plot legend
    plt.legend(title=leg_title, loc=loc, labels=leg_labels)

    return plot_bar
```

```python
plot_bar(
    x='Year', y='Value',
    hue='MajorTradingPartner_1', 
    df=df1_gpd.get_group('United States of America'),
    main_title='USA', xlabel='Year', ylabel='Percentage of Imports_Exports',
    leg_title='Major trading partners'
)
```
