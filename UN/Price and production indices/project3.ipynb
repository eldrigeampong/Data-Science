# Data Visualization

```python
# helper function to plot line charts
def plot_line(x, y, hue, df, title=None, xlabel=None, ylabel=None):
    # initiate matplot figure
    fig, ax = plt.subplots(figsize=(14, 12))
    fig.autofmt_xdate()

    # plot line chart
    sns.barplot(
        x=x, y=y,
        hue=hue,
        data=df)

    # label plot axis
    ax.set(title=title, xlabel=xlabel, ylabel=ylabel)

    return plot_line
```

```python
plot_line(
    x='Year', y='Value',
    hue='IndexIndustrialProduction', 
    df=df1_gpd.get_group('Albania'), 
    xlabel='Year', ylabel='Index of industrial production'
)
```
